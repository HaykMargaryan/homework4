```python
import requests
from BeautifulSoup import *
import json
```

```python
url ='https://www.youtube.com/feed/trending'
re=requests.get(url)
data=re.text
soup=BeautifulSoup(data)
X=soup.findAll('ul', attrs={'class':"expanded-shelf-content-list has-multiple-items"})
Data_movies=[]
for i in range(len(X)):
    x=X[i].findAll('h3')
    y=X[i].findAll('div', attrs={'class':"yt-lockup-byline "})
    z=X[i].findAll('ul',attrs={'class':"yt-lockup-meta-info"})
    for j in range(len(x)):
        t={}
        link=x[i].find('a').get('href')
        title=x[i].find('a').get('title')
        dur=str(x[i].find('span').text)
        user=y[i].find('a').text
        channel=y[i].find('a').get('href')
        views=z[i].findAll('li')[1].text
        t['link'] = "https://www.youtube.com"+str(link)
        t['video'] = title
        t['duration'] = dur[11:-1]
        t['user_name'] = user
        t['channel'] = "https://www.youtube.com"+str(channel)
        t['views']  = views
        Data_movies.append(t)
    print Data_movies           
        
```

```python
import urllib2 as u2
from lxml import html
from lxml.html import document_fromstring
url ='https://www.youtube.com/feed/trending'
response = u2.urlopen(url)
page = response.read()
tree = document_fromstring(page)
```

```python
data=tree.xpath('//ul[@class="expanded-shelf-content-list has-multiple-items"]')
data_movies_1=[]
for i in range(len(data)):
    x=data[i].cssselect('h3')
    y=data[i].cssselect('[class="yt-lockup-byline "]')
    z=data[i].cssselect('[class="yt-lockup-meta-info"]')
    for i in range(len(x)):
        t={}
        link=x[i].cssselect('a')[0].get('href')
        title=x[i].cssselect('a')[0].text_content()
        dur=str(x[i].cssselect('span')[0].text_content())
        user=y[i].cssselect('a')[0].text_content()
        channel=y[i].cssselect('a')[0].get("href")
        views=z[i].cssselect('li')[1].text_content()
        t['link'] = "https://www.youtube.com"+str(link)
        t['video'] = title
        t['duration'] = dur[12:-1]
        t['user_name'] = user
        t['channel'] = "https://www.youtube.com"+str(channel)
        t['views']  = views
        data_movies_1.append(t)
```

```python
pprint(data_movies_1)
```

```python
len(data_movies_1)
```

```python
with open('videos_xml.json', 'w') as output:
    json.dump(data_movies_1, output, sort_keys = True, indent = 4)
```

```python
import unicodecsv as csv
with open('trending_videos_xml.csv', 'w') as csv_file:
    for i in data_movies_1:
            writer = csv.writer(csv_file)
            for key, value in i.items():
                writer.writerow([key, value])
            writer.writerow([])
```
